```
# imports
import numpy as np
import struct
import glob
from PIL import Image
from itertools import chain
```

```
# folder location 
PATH_DIR = '/content/drive/My Drive/'
```

```
# read training dataset (IDX file format)

with open(PATH_DIR + 'train-images.idx3-ubyte','rb') as f:
    magic, size = struct.unpack(">II", f.read(8))
    nrows, ncols = struct.unpack(">II", f.read(8))
    data_train = np.fromfile(f, dtype=np.dtype(np.uint8).newbyteorder('>'))
    data_train = data_train.reshape((size, nrows, ncols))
```
